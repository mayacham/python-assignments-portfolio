```python
### part a

import sympy as sp
import numpy as np
import math 

x, y = sp.symbols('x y')

f = sp.exp(x) * sp.sin(y) + y**3

df_dx = sp.diff (f, x)

df_dy = sp.diff (f, y)

print ("df_dx:", df_dx)
print ("df_dy:", df_dy)
```

```python
#### part b

x, y = sp.symbols('x y')

g = x**2 * y + x * y**2
p =[1, -1]


# Calculate the derivatives
dg_dx = sp.diff(g, x)
dg_dy = sp.diff(g, y)

# Evaluate the derivatives at the point (1, -1)
dg_dx_value = dg_dx.subs({x: 1, y: -1})
dg_dy_value = dg_dy.subs({x: 1, y: -1})


# Create grad array with the evaluated derivatives
grad = np.array([dg_dx_value, dg_dy_value])
print("grad:", grad)

magnitude = sp.sqrt(dg_dx_value**2 + dg_dy_value**2)

print("magnitude:", magnitude)
```

```python
#### part c

x, y = sp.symbols('x y')
h = sp.log(x**2 + y**2)

dh_dx = sp.diff(h, x)
dh_dy = sp.diff(h, y)

d2h_dx2 = sp.diff(dh_dx, x)
d2h_dy2 = sp.diff(dh_dy, y)
d2h_dxy = sp.diff(dh_dx, y)
d2h_dyx = sp.diff(dh_dy, x)

print ("d2h_dx2:", d2h_dx2)
print ("d2h_dy2:", d2h_dy2)
print ("d2h_dxy:", d2h_dxy)
print ("d2h_dyx:", d2h_dyx)
```

```python
## part d

import matplotlib.pyplot as plt

x, y = sp.symbols('x y')
j = x**3 - 3*x*y + y**3

j_func = sp.lambdify((x, y), j, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = j_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $j(x, y) = x^3 - 3xy + y^3$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()


x, y = sp.symbols('x y')
k = y**2 - x**2

k_func = sp.lambdify((x, y), k, 'numpy')

x_vals = np.linspace(-3, 3, 400)
y_vals = np.linspace(-3, 3, 400)
X, Y = np.meshgrid(x_vals, y_vals)
Z = k_func(X, Y)

plt.contourf(X, Y, Z, levels=50, cmap='viridis')
plt.colorbar()
plt.title('Contour plot of $k(x, y) = y^2 - x^2$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

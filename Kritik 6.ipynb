```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate

# Given variables:
average_height = 171
variance = 7.1**2


def normal_density(mean, variance, x):
    
    #### normal_density=(1/(sqrt(2*np.pi*variance**2)))*exp(1)**((-(x-variance)**2)/2*variance**2)
    ### very long and easy to make mistakes, so simplify by
    ### making the fraction coefficient and exponent into variables
    
    exponent = -((x - mean) ** 2) / (2 * variance)
    cfcnt = 1 / np.sqrt(2 * np.pi * variance)
    
    return cfcnt * np.exp(exponent)

def plot_normal_density(mean, variance, label):
    ####plotting the function to visualize it
    x_values = np.linspace(mean - 3 * np.sqrt(variance), mean + 3 * np.sqrt(variance), 100)
    ####common practice to get a domain around the mean - gets almost all values
    y_values = normal_density(mean, variance, x_values)
    plt.plot(x_values, y_values, label=label)

def integration(mean, variance, a, b):
    result, _ = integrate.quad(lambda x: normal_density(mean, variance, x), a, b)
    return result
    


# Plot normal density function for given μ and σ**2
plot_normal_density(171, 7.1, label='μ=171, σ^2=7.1**2')

plt.xlabel('Height (cm)')
plt.ylabel('Probability Density')
plt.show()

# Calculate the probability of a random man having a height between 162cm and 190cm
probability = integration(average_height, variance, 162, 190)
print(f"The probability of a randomly chosen man having a height between 162cm and 190cm is: {probability:.4f}")
```
